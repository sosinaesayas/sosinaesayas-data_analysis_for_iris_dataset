```python
threshold = len(data) * 0.02  # Drop columns with more than 50% null values
data_cleaned = data.dropna(axis=1, thresh=threshold)
```

```python
data = data.drop(["sepal_length", "sepal_width"], axis = 1)
# data = data.drop(columns = ["sepal_length", "sepal_width"], inplace = True)  is equal to the above
```

```python
data.head()
```

```python
#Univariant Analysis
# univariate plots
fig, ax=plt.subplots(1,2, figsize=(20, 6))
sns.histplot(data['petal_length'], ax=ax[0])
ax[0].set_title("petal_length")
sns.histplot(data['petal_width'], ax=ax[1])
ax[1].set_title("petal_width")
plt.ylim(0, 2500)

```

```python
# bivariant analysis
a4_dims = (20, 7)
fig, axs = plt.subplots(ncols=3, figsize=a4_dims)
sns.scatterplot(x="petal_length", y="petal_width", ax=axs[0], data=data)
sns.scatterplot(x="petal_length", y="species", ax=axs[1], data=data)
sns.scatterplot(x="petal_width", y="species", ax=axs[2], data=data)

plt.show()
```
